```python
# Setting release_date column as index

df_tracks.set_index("release_date", inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
# Changing duration from milliseconds to seconds

df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace=True, axis=1)
```

```python
df_tracks.head()
```

### Correlation Matrix Between different variables

```python
corr_df = df_tracks.drop(['key','mode','explicit'], axis=1).corr(method='pearson')
corr_df
```

### Correlation HeatMap Between Different Variables

```python
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, linewidths=0.5, linecolor='Black')
heatmap.set_title("Correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
plt.show()
```

```python
sample_df = df_tracks.sample(int(0.006*len(df_tracks)))
print(len(sample_df))
```

### Loudness vs Energy Correlation

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, x='acousticness', y='loudness', color='c').set(title = 'Loudness vs Acousticness Correlation')
```

### Popularity vs Energy Correlation

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

### Number of songs released per year

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title="Number of songs per year")
```

### Song duration over the years

```python
total_dr = df_tracks['duration']
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x=years, y=total_dr, ax=ax, errwidth=False).set(title="Year vs Duration")
plt.xticks(rotation=90)
plt.show()
```

### Importing dataset containing genre data

```python
df_genre = pd.read_csv('H:/Downloads/Spotify data/SpotifyFeatures.csv')
```

```python
df_genre.head()
```

### Most Popular Genres

```python
plt.figure(figsize=(14,6))
df_genre.groupby('genre')['popularity'].sum().sort_values(ascending=False).plot(kind='bar')
```

### Average duration of songs in different genres

```python
plt.figure(figsize=(15,7))
plt.title("Duration of songs in different genres")
sns.color_palette("rocket", as_cmap=True)
sns.barplot(x='genre', y = 'duration_ms', data=df_genre)
plt.xticks(rotation=90)
plt.show()
```

## Conclusion
- With rise of the internet and multiple platforms available for music artists to release their music, more and more songs are released each year.
- Songs that are energetic and loud have higher probability of been popular.
- Duration of songs is declining year by year recently, which may be because the attention span of humans has reduced over the years. So, artists prefer making shorter songs.
- Pop is the most popular genre followed by Rap and Rock.
- The songs made for small kids are very short in duration because kids have very short attention span compared to adults.
